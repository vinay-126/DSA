From correlation table, we can get various insights like:

So, we can safely say that even if we remove other columns such as positioning, penalties, curve, ball control, free kick accuracy, short & long passing, we can get them by help of only dribbling data..

```python
figure,axis=plt.subplots(2,2)

drib=list(df.Dribbling)
vol=list(df.Volleys)
res5=drib[1:100]
res6=vol[1:100]

axis[0,0].plot(res5,res6)
axis[0,0].set_title('Volleys vs Dribbling')

pos=list(df.Positioning)
res7=pos[1:100]

axis[0,1].plot(res5,res7)
axis[0,1].set_title('Positioning vs Dribbling')

curve=list(df.Curve)
res8=curve[1:100]

axis[1,0].plot(res5,res8)
axis[1,0].set_title('Curve vs Dribbling')

con=list(df.Crossing)
res9=con[1:100]

axis[1,1].plot(res5,res9)
axis[1,1].set_title('Crossing vs Dribbling')
plt.show()
```

We can confirm our data about dribbling from above cases also.
